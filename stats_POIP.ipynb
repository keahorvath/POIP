```python
name = 'https://raw.githubusercontent.com/keahorvath/POIP/refs/heads/main/warehouse_big_category/product_circuit.txt'
n_produits = pd.read_csv(name, header=None, nrows=1).iloc[0, 0]
fam_big_category = pd.read_csv(name, header=None, skiprows=1, names=["famille"])

fam_big_category["produit"] = fam_big_category.index
fam_big_category = fam_big_category[["produit", "famille"]]

fam_big_category.head()
```

```python
name = "https://raw.githubusercontent.com/keahorvath/POIP/refs/heads/main/warehouse_big_category/orders.txt"
lines = pd.read_csv(name, header=None).squeeze().tolist()

commandes_big_category = []
for line in lines[1:]:
    vals = list(map(int, line.split()))
    n = vals[0]
    produits = vals[1:1+n]
    commandes_big_category.append(produits)

commandes_big_category = pd.DataFrame({"produits": commandes_big_category})
commandes_big_category["order"] = commandes_big_category.index
commandes_big_category = commandes_big_category[["order", "produits"]]

commandes_big_category.head()
```

<h2 style="padding:16px; font-size:40px; color:#FFF; background:#07D">Nb de circuits diff par commande</h2>

<h2 style="padding:16px; font-size:20px; color:#FFF; background:#07D">Big category</h2>

```python
commandes_big_category["nb_circuits"] = commandes_big_category["produits"].apply(
    lambda prods: len(set(fam_big_category.set_index("produit").loc[prods, "famille"]))
)
commandes_big_category.head()
```

```python
print("min :", commandes_big_category.nb_circuits.min(), "\nmax :", commandes_big_category.nb_circuits.max(),
     "\nmédiane :", commandes_big_category.nb_circuits.median(), "\nmoyenne :", commandes_big_category.nb_circuits.mean())
```

```python
freq_big_category = commandes_big_category["nb_circuits"].value_counts().sort_index()
freq_big_category.head()
```

```python
plt.bar(freq_big_category.index, freq_big_category.values)

plt.xlabel("Nombre de circuits")
plt.ylabel("Nombre de commandes")
# plt.title("Distribution du nombre de circuits par commande")

plt.show()
```
